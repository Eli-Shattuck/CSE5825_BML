```python
from beschess.load import load_puzzle
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%load_ext autoreload
%autoreload 2
```

```python
# PuzzleId,FEN,Moves,Rating,RatingDeviation,Popularity,NbPlays,Themes,GameUrl,OpeningTags
RATING_IDX = 3
THEMES_IDX = 7
```

```python
puzzles = load_puzzle.puzzle_csv_iterator(r'../data/puzzle.csv')
themes_count = Counter()
ratings_count = Counter()
for puzzle in puzzles:
    themes_count.update(puzzle[THEMES_IDX].split(' '))
    ratings_count.update([puzzle[RATING_IDX]])


themes_count, ratings_count
```

```python
themes_sorted = dict(sorted(themes_count.items(), key=lambda item: item[1], reverse=False))

plt.figure(figsize=(6, 10))
plt.barh(themes_sorted.keys(), themes_sorted.values(), color='skyblue')
plt.xticks(rotation=90)
plt.tight_layout()
plt.margins(y=0)
plt.xscale('log')
plt.show()
```

```python
ratings_sorted = dict(sorted(ratings_count.items(), key=lambda item: int(item[0]), reverse=False))

plt.figure(figsize=(10, 5))
# n, bins, _ = plt.hist(ratings_sorted.keys(), weights=list(ratings_sorted.values()), bins=ratings_sorted.keys(), color='skyblue', width=1)
plt.scatter(ratings_sorted.keys(), ratings_sorted.values(), color='skyblue')
plt.xticks(list(ratings_sorted.keys())[::100], rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
n, bins, _ = plt.hist(ratings_sorted.keys(), weights=list(ratings_sorted.values()), bins=50, color='skyblue')
plt.xticks(bins, rotation=90)
plt.show()
```
